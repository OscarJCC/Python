$$f(x) = x$$
$$g(x) = x^2$$
$$h(x) = x^3$$

```python
x = np.linspace(1,4,1001)
print(x)

fig, ax = plt.subplots()
ax.grid()
ax.set_title(r"$Grafica$ $simple$ $f(x)$, $g(x)$, $h(x)$")
ax.plot(x,x,label = r"$f(x) = x$")    #f(x)
ax.plot(x,x**2,label = r"$g(x) = x^2$") #g(x)
ax.plot(x,x**3,label = r"$h(x) = x^3$") #h(x)
ax.set_xlabel("Eje x")
ax.set_ylabel("Eje y")
ax.text(2,30,r"$\int_{a}^{b}$",fontsize = 20)
ax.legend()
plt.show()
```

```python
x = np.linspace(-2*np.pi,2*np.pi,1001)
print(x)

f = np.sin(x)
g = np.cos(x)

fig, ax = plt.subplots()
ax.grid()
ax.set_title("Funciones trigonometricas")
ax.plot(x,f,color = "red",label = r"$\sin(x)$")
ax.plot(x,g,color = "blue",label = r"$\cos(x)$")
ax.set_xlabel("Eje x")
ax.set_ylabel("Eje y")
ax.set_xlim(min(x),max(x))
ax.set_ylim(min(f),max(f))
ax.legend()
plt.show()
```
