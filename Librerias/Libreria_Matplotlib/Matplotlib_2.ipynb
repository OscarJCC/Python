```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-2*np.pi, 2*np.pi,1001)

f = np.sin(x)
g = np.cos(x)
```

```python
fig, (ax1,ax2,ax3,ax4) = plt.subplots(1,4,figsize = (15,7.5))

ax1.set_title("Grafica 1")
ax1.plot(x,f)
ax1.grid()

ax2.plot(x,g)
ax2.grid()

ax3.plot(x,x**2)
ax3.grid()

ax4.plot(x,x**3-5,color = "orange")
ax4.grid

plt.show()
```

```python
t = np.linspace(-1,1,21)

plt.style.use("dark_background")

fig, ax = plt.subplots()

ax. plot(t,np.exp(t),marker = "o")
ax.grid(True)
plt.show()
```
