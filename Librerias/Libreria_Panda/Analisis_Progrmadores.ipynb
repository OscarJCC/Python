```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
results = pd.read_csv("./Data/survey_results_public.csv", index_col = "ResponseId")
resultsSchema = pd.read_csv("./Data/survey_results_schema.csv")
```

```python
results.head(5)
```

```python
languages = "LanguageHaveWorkedWith"
salary = "ConvertedCompYearly"

#results = results.rename(columns={languages: "Languages", salary: "Salary"}) # Podemos renombrar las columnas
results.rename(columns={languages: "Languages", salary: "Salary"}, inplace=True) # con inplace=True hacermos que la misma variable de datos se remplasada

results[["Languages","Salary"]]
```

```python
results.dropna(subset=["Languages","Salary"], inplace = True) #Permite eliminar todos los datos que tengan NaN
results[["Languages","Salary"]]
```

```python
results = results.loc[:,["Country","Salary","Languages"]].sort_values(by="Salary") # con .loc() redefinimos results solo tomando las columnas deseadas y ordenamos segun el salario
results
```

```python
filtro = (results["Salary"] >= 10000) & (results["Salary"] <= 3e6)
results = results[filtro] #Estmos filtrando los datos con salarios no creibles
results
```

```python
lg = results["Languages"]
allLanguages = []

for i in lg:
     for j in i.split(";"):  #.split nos devuelve una lista con los elementos de un string que estan separados por algun simbolo en este caso un ; 
          if j not in allLanguages:
               allLanguages.append(j)
          
#allLanguages = list(set(allLanguages)) # set nos permite crea un conjunto de elementos de una lista con valores unicos sin que se repitan
allLanguages.sort() #sort nos ordena alfabeticamente
allLanguages
```

```python
meanSalary = pd.DataFrame(data = np.zeros(len(allLanguages)),index = allLanguages, columns=["Salary"])
meanSalary
```

```python
import re

for i in allLanguages:
     mask = results["Languages"].str.contains(re.escape(i)) #con .str.contains buscamos en un str si esta el argumento que damos
     average = results.loc[mask,"Salary"].mean()
     meanSalary.loc[i,"Salary"] = average

meanSalary
```

```python
%matplotlib qt5

meanSalary=meanSalary.sort_values(by="Salary")

plt.style.use("dark_background")
plt.subplots(facecolor = "black", layout = "constrained")
plt.barh(meanSalary.index,meanSalary["Salary"], color = "blue")
plt.vlines(meanSalary["Salary"].mean(),0,len(meanSalary["Salary"]), colors="red", label="Mean Salary")
plt.title("Programming Languages Yearly Compensation")
plt.xlabel("Mean Salary in $")
plt.ylabel("Programming languages")
plt.gca().xaxis.set_major_formatter('${x:1.0f}')
plt.grid(False)
plt.legend()
plt.tight_layout()
plt.show()

```
